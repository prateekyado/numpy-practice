```python
#4. Count how many values are greater than the mean of the array.
arr4 = np.array([5,10,15,20,25,30])
```

```python
mean = arr4.mean()
mean
```

```python
count = np.sum(arr4>mean)
count
```

```python
matrix = np.array([[1, 9, 10],
                   [4, 2, 6]])

np.argmax(matrix)       
np.unravel_index(np.argmax(matrix), matrix.shape)
```
