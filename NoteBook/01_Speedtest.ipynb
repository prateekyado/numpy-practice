```python
#using python list
import time

size = 1_000_000_0
l1 = list(range(size))
l2 = list(range(size))

start = time.time()
add = [x+y for x,y in zip(l1,l2)]
print(add[0:20])
end = time.time()

print(end - start)
```

```python
#using numpy array
import numpy as np
import time

size = 1_000_000_0
l1 = np.array(list(range(size)))
l2 = np.array(list(range(size)))

start = time.time()
add = l1 + l2
print(add[0:20])
end = time.time()

print(end - start)
```
